# Import Modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import LabelEncoder
from pycaret.classification import * 
%matplotlib inline
warnings.filterwarnings('ignore')
```

# Load the Dataset

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# delete unnecessary columns
df = df.drop(columns=['id', 'Unnamed: 32'], axis=1)
```

```python
# statistical info
df.describe()
```

```python
# datatype info
df.info()
```

# Exploratory Data Analysis

```python
# Count the occurrences of each category
diagnosis_counts = df['diagnosis'].value_counts()

# Create a DataFrame from the counts
diagnosis_counts_df = pd.DataFrame({'Diagnosis': diagnosis_counts.index, 'Count': diagnosis_counts.values})

# Create the countplot
sns.barplot(x='Diagnosis', y='Count', data=diagnosis_counts_df)
```

```python
# Create dist plot
fig, ax = plt.subplots(ncols=5, nrows=6, figsize=(20, 20))
index = 0
ax = ax.flatten()

for col in df.columns:
    if np.issubdtype(df[col].dtype, np.number):  # Check if the column contains numeric data
        sns.distplot(df[col], ax=ax[index])
        index += 1

plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
# Create box plot
fig, ax = plt.subplots(ncols=6, nrows=5, figsize=(20, 20))
index = 0
ax = ax.flatten()

for col in df.columns:
    if np.issubdtype(df[col].dtype, np.number):  # Check if the column contains numeric data
        sns.boxplot(y=col, data=df, ax=ax[index])
        index += 1

plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

# Create and Train the Model

```python
# setup the data
clf = setup(df, target='diagnosis')
```

```python
# train and test the models
compare_models()
```

```python
from pycaret.classification import create_model

model = create_model('rf')
```

```python
# hyperparameter tuning
best_model = tune_model(model)
```

```python
evaluate_model(best_model)
```

```python
# plot the results
plot_model(estimator=best_model, plot='confusion_matrix')
```
